```python
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt_tab')
import warnings

warnings.filterwarnings(action='ignore')
```

```python
text = "Hi everyone. I am playing footbal. I thought I should consider taking care of my health so I basically though to play it regularly so that all th unpleasant vibes goes out of my thinking."
```

```python
data = []

for i in sent_tokenize(text):
    temp = []
    for j in word_tokenize(i):
        temp.append(j.lower())

    data.append(temp)
```

```python
model1 = Word2Vec(data, min_count=1, vector_size=100, window=5, sg=0) #CBOW
model2 = Word2Vec(data, min_count=1, vector_size=100, window=5, sg=1) #skip gram
```

```python
model1.wv.similarity('health', 'care')
```

```python
model1.wv.similarity('footbal', 'playing')
```
